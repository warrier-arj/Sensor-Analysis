```python
import bagpy
from bagpy import bagreader
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
import numpy as np

## data stored to b
b = bagreader('../data/stationary_data.bag')
b2 = bagreader('../data/walking_data.bag')
#shows a table of all topics with information of how many msgs in them
print(b.topic_table)
```

```python
# saving to csv
st= b.message_by_topic(topic='/gps')
walker = b2.message_by_topic(topic='/gps')
stand = pd.read_csv(st)
walk = pd.read_csv(walker)
walk
```

```python
# plotting
plt.plot(stand['UTM_Northing'],stand['UTM_Easting'])
plt.ylabel('UTM Easting(in m)')
plt.xlabel('UTM Northing(in m)')
plt.title('Easting vs Northing for stationary.')
```

```python
plt.plot(stand['Header.stamp.secs'],stand['Altitude'])
plt.ylabel('Altitude(in m)')
plt.xlabel('Time Stamp(in sec)')
plt.title('Altitude against time for a stationary point.')
```

```python
import statistics
m = statistics.mean(stand['UTM_Northing'])
print("Mean of Northing while stationary.")
print(m)

mn = statistics.mean(stand['UTM_Easting'])
print("Mean of Easting while stationary.")
print(mn)
```

```python
plt.plot(stand['UTM_Easting'],stand['Header.stamp.secs'])
plt.xlabel('Easting(in m)')
plt.ylabel('Time Stamp(in sec)')
plt.title('Easting against time for stationary.')
```

```python
plt.plot(stand['UTM_Northing'],stand['Header.stamp.secs'])
plt.xlabel('Northing(in m)')
plt.ylabel('Time Stamp(in sec)')
plt.title('Northing against time for stationary .')
```
